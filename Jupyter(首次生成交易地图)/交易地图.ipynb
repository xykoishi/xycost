```python
import re
import os
import time
from numpy import mean

deal = []
site = []
site_lat=[]
site_lng=[]

with open(file,"r") as f_o:
    lines = f_o.read()
    lines = lines.split('\n')

    for line in lines:
        if "#" in line:
            #print(line)
            line = line.split("#")
            site.append([line[1],line[2]])   # åªä¿å­˜åæ ‡ï¼Œå¯ç”¨äºç”»çƒ­åŠ›å›¾
            line[0] = line[0].split(',')
            line[0] = str(line[0][2])+" ("+str(line[0][1])+"): "+str(line[0][3])+"  ["+str(line[0][0])+"]"
            deal.append(line)
            
            site_lat.append(round(float(line[1]),2))
            site_lng.append(round(float(line[2]),2))
            
site_lat=list(set(site_lat))
site_lng=list(set(site_lng))
center=[mean(site_lat),mean(site_lng)]
#print(deal)
```

```python
import folium
from folium.plugins import MarkerCluster
from folium.plugins import HeatMap
#
m = folium.Map(center, zoom_start=10, tiles='CartoDB',attr='positron')

#m = folium.Map(center, tiles= 'https://wprd01.is.autonavi.com/appmaptile?x={x}&y={y}&z={z}&lang=zh_cn&size=1&scl=1&style=7',
#           attr='é«˜å¾·-å¸¸è§„å›¾', zoom_start=10)

# è°·æ­Œåœ°å›¾ å›½å¤–æ¥å£
#m = folium.Map(center, tiles='https://mt.google.com/vt/lyrs=m&x={x}&y={y}&z={z}',
#               attr='default', zoom_start=5)

# è°·æ­Œåœ°å›¾ å›½å†…æ¥å£ã€å›½å¢ƒæ­£ç¡®ã€‘
#m = folium.Map(center, tiles='http://www.google.cn/maps/vt?pb=!1m5!1m4!1i{z}!2i{x}!3i{y}!4i256!2m3!1e0!2sm!3i342009817!3m9!2sen-US!3sCN!5e18!12m1!1e47!12m3!1e37!2m1!1ssmartmaps!4e0&token=32965',
#               attr='default', zoom_start=5)

style = 'CartoDB_Voyager'  # Jawg_Streets' # 'CartoDB_Voyager'

if style == 'Jawg_Streets':   # è™½ç„¶æ˜¯æœ€æ»¡æ„çš„æ ·å¼ï¼Œä½†åŠ è½½é€Ÿåº¦å¤ªæ…¢äº†
    URL = 'https://tile.jawg.io/jawg-streets/{z}/{x}/{y}{r}.png?access-token=B0VlJHdtlgJXigsV4NpHew7W7IBM7rR2N619R9zNr5lhtZ2YAav240ffGdNLetmY'
    attribution = '<a href=\"https://www.jawg.io\" target=\"_blank\">&copy; Jawg</a> - <a href=\"https://www.openstreetmap.org\" target=\"_blank\">&copy; OpenStreetMap</a>&nbsp;contributors'

if style == 'CartoDB_Voyager':
    URL = 'https://{s}.basemaps.cartocdn.com/rastertiles/voyager/{z}/{x}/{y}{r}.png'
    attribution = '&copy; <a href="https://www.openstreetmap.org/copyright">OpenStreetMap</a> contributors &copy; <a href="https://carto.com/attributions">CARTO</a>'

m = folium.Map(center, tiles=URL, attr=attribution, 
               zoom_start=5)
```

```python

# create a mark cluster object
marker_cluster = MarkerCluster().add_to(m)

# add data point to the mark cluster
for label, lat, lng in deal:
    folium.Marker(
        location=[lat,lng],
        icon=None,
        popup=folium.Popup(label,max_width=500),
    ).add_to(marker_cluster)

m.add_child(marker_cluster)

#HeatMap(deal).add_to(m)
```
